```python
raw_data_fbref = pd.read_csv("../../data/raw/raw_data_fbref.csv")
raw_data_fbref.head()
```

#### Change countries' name format

```python
# Countries names format and their frequency
count = raw_data_fbref["Country"].value_counts()
count
```

**There are some letters before the capital countries' abbreviation name which is the flags on website. Therefore, dropping these letters is necessary**

```python
# get rid of the first word of countries name
raw_data_fbref['Country'] = raw_data_fbref['Country'].str.split().str[1]
count = raw_data_fbref["Country"].value_counts()
count
```

#### Drop column 'Matches'

```python
raw_data_fbref['Matches'].unique()
```

**Column 'Matches' only has Matches value. In fact, on website this is a hyperlink to another match so dropping this column is necessary**

```python
drop_raw_data_fbref = raw_data_fbref.drop(['Matches'],axis = 1)
```

#### Adjusting season column format

```python
season_count = drop_raw_data_fbref['Season'].value_counts()
print(season_count)
print(f"total: {season_count.sum()}")
```

**Data of players currently playing in premier league may include their previous records from other leagues in which season format is different. It is neccessary to adjust this for later stages.**

```python
season_data_fbref = drop_raw_data_fbref.copy()
def format_season(season):
    if pd.notna(season) and '-' in season:
        return season
    elif pd.notna(season):
        return f"{season}-{int(season) + 1}"
    else:
        return season
season_data_fbref['Season']= season_data_fbref['Season'].apply(format_season)
```

```python
season_count = season_data_fbref['Season'].value_counts()
print(season_count)
print(f"total: {season_count.sum()}")
```

#### Checking and adjusting columns data type 

```python
season_data_fbref.dtypes
```

**Overall, data type of most fields is correct, except for minutes played (Min) and some basic football statistic like goals and assists should be int instead of float but there are some missing value filled with NaN so these columns will always be float**

```python
# format min column
type_data_fbref = season_data_fbref.copy()
def format_min(min):
     if pd.notna(min):
        return int(min.replace(',', ''))
     else:
        return min
    
type_data_fbref['Min'] = type_data_fbref['Min'].apply(format_min)
```

### 2.2. Transfermarkt preprocessing

```python
raw_data_trans = pd.read_csv("../../data/raw/raw_data_trans.csv")
raw_data_trans.head()
```

#### Adjusting Season Format

```python
seas_count = raw_data_trans['Season'].value_counts()
print(seas_count)
print(f"Total: {seas_count.sum()}")
```

**We need to change season format to match the one of fbref for later stage**

```python
season_data_trans = raw_data_trans.copy()
def format_trans_season (season):
    if pd.notna(season):
        year = season.split('/')
        return f"20{year[0]}-20{year[1]}"
    else:
        return season
season_data_trans['Season'] = season_data_trans['Season'].apply(format_trans_season)
seas_count = season_data_trans['Season'].value_counts()
print(seas_count)
print(f"Total: {seas_count.sum()}")
```

**Adjusting Days format to int**

```python
days_data_trans  = season_data_trans.copy()
days_data_trans['Days'] = days_data_trans['Days'].str.split().str[0].astype('int64')
days_data_trans['Days'].dtypes
```

## 3. Merging data

**Merging 2 dataframes can help to attach additional information on injured player for later stage**

There is one problem that the name of players of 2 dataframes is not always similar, therefore we need an assisting package called fuzzywuzzy to evaluate how 2 strings are indentical to merge

```python
#!pip install fuzzywuzzy
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

# create a dictionary for matching names of 2 Name columns of 2 dataframes
name_matches = {}
for name in type_data_fbref['Name'].unique():
    best_match = process.extractOne(name, days_data_trans['Name'].unique())
    name_matches[name] = best_match
type_data_fbref_copy = type_data_fbref.copy()
type_data_fbref_copy['Name'] = type_data_fbref_copy['Name'].map(name_matches)

# Merge based on both 'Name' and 'Season'
merged_df = pd.merge(type_data_fbref_copy, days_data_trans , left_on=['Name', 'Season'], right_on=['Name', 'Season'], how='left')
```

```python
def extract_first_name(value):
    return value[0]
merged_df['Name'] = merged_df['Name'].apply(extract_first_name)
merged_df
```
